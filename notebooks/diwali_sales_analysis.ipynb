```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diwali_data = pd.read_csv("../dataset/diwali_sales_data.csv",encoding="cp1252")
```

Here are some data analysis questions you can explore with this **Diwali Sales Dataset**:

1. **Demographics & Customer Insights**

   * Which **age group** has the highest number of purchases?
   * Is there a significant difference in total sales between **male** and **female** customers?
   * Which **occupation** group contributes the most to total sales?
   * What is the distribution of customers across different **zones** in India?

2. **Sales Performance**

   * Which **state** generated the highest total sales amount during Diwali?
   * What is the **average purchase amount per order** for each product category?
   * Which **product category** has the highest total sales and number of orders?
   * Is there a correlation between the **number of orders** and the **amount spent**?

3. **Behavioral Insights**

   * Do **married customers** spend more than unmarried customers?
   * Which age group tends to place the **largest orders (by amount)**?
   * What is the **most popular product** by order count and total revenue?

4. **Advanced Analysis**

   * Can you segment customers into **high spenders** and **low spenders**, and find their key demographic differences?
   * Using the `Occupation` and `State` fields, can you identify the **best market segment** for future campaigns? 

```python
purchases_age_data = diwali_data.groupby("Age Group").aggregate('Amount').sum()
```

```python
purchases_age_data
```
