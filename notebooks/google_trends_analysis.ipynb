### Google trends analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
interest_over_time = pytrends.interest_over_time()
```

```python
interest_over_time
```

```python
kw_list = ["Cloud Computing"]
pytrends.build_payload(kw_list, cat=0, timeframe='2024-01-01 2024-02-01', geo='', gprop='')
data = pytrends.interest_over_time()
data = data.sort_values(by="Cloud Computing", ascending = False)
data = data.head(10)
```

```python
data
```

```python
data = pytrends.interest_over_time()
data = data.sort_values(by="Cloud Computing", ascending = False)
data = data.head(10)
data.reset_index()
```

```python
data = pytrends.interest_by_region()
data = data.sort_values(by="Cloud Computing", 
                        ascending = False)
data = data.head(10)
```

```python
data
```

```python
data.reset_index().plot(x='geoName', y='Cloud Computing',
                        figsize=(10,5), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```
